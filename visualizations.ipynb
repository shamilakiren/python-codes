Exercise 1: Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties:
● X label name = Month Number
● Y label name = Total profit

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Extract month number and total profit
month = df['month_number']
total_profit = df['total_profit']

# Plot the line chart
plt.figure(figsize=(10,5))
plt.plot(month, total_profit, marker='o', color='blue', linewidth=2)

# Add labels and title
plt.xlabel("MonthNumber")
plt.ylabel("Totalprofit")
plt.title("Total Profit per Month")

# Show grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:
● Line Style dotted and Line-color should be red
● Show legend at the lower right location.
● X label name = Month Number
● Y label name = Sold units number
● Add a circle marker.
● Line marker color as read
● Line width should be 3

```python
# Extract month number and total profit
month = df['month_number']
total_profit = df['total_profit']

# Plot with required style properties
plt.figure(figsize=(10,5))
plt.plot(
    month,
    total_profit,
    linestyle='dotted',      # dotted line
    color='red',             # line color red
    marker='o',              # circle marker
    markerfacecolor='red',   # marker color red
    linewidth=3,             # line width 3
    label='Total Profit'     # legend label
)

# Labels
plt.xlabel("MonthNumber")
plt.ylabel("Soldunits number")

# Legend at lower right
plt.legend(loc='lower right')

# Grid for readability
plt.grid(True)

# Show plot
plt.show()
```

Exercise 3: Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline
for each product).

```python
# Extract month numbers
month = df['month_number']

# Plot each product as a separate line
plt.figure(figsize=(12,6))

plt.plot(month, df['facecream'], marker='o', linewidth=2, label='Face Cream')
plt.plot(month, df['facewash'], marker='o', linewidth=2, label='Face Wash')
plt.plot(month, df['toothpaste'], marker='o', linewidth=2, label='Toothpaste')
plt.plot(month, df['bathingsoap'], marker='o', linewidth=2, label='Bathing Soap')
plt.plot(month, df['shampoo'], marker='o', linewidth=2, label='Shampoo')
plt.plot(month, df['moisturizer'], marker='o', linewidth=2, label='Moisturizer')

# Labels and title
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("Monthly Sales Data for Each Product")

# Show legend
plt.legend(loc='upper left')

# Grid for readability
plt.grid(True)

# Display plot
plt.show()
```

Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in
the plot. gridline style should “–“.

```python
# Extract month number and toothpaste sales
month = df['month_number']
toothpaste_sales = df['toothpaste']

# Create scatter plot
plt.figure(figsize=(10,5))
plt.scatter(month, toothpaste_sales, color='blue', marker='o')

# Labels and title
plt.xlabel("Month Number")
plt.ylabel("Toothpaste Units Sold")
plt.title("Monthly Toothpaste Sales")

# Add grid with dashed style
plt.grid(True, linestyle='--')

# Show plot
plt.show()
```

Exercise 5: Read face cream and facewash product sales data and show it using the bar chart.

```python
# Extract month numbers and product sales
month = df['month_number']
facecream = df['facecream']
facewash = df['facewash']

# Set bar width
bar_width = 0.4

# Create bar positions
x1 = month
x2 = month + bar_width

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(x1, facecream, width=bar_width, label='Face Cream', color='skyblue')
plt.bar(x2, facewash, width=bar_width, label='Face Wash', color='orange')

# Labels and title
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("Monthly Sales of Face Cream and Face Wash")

# Add legend
plt.legend()

# Show plot
plt.show()
```

Exercise 7: Read the total profit of each month and show it using the histogram to see the most common
profit ranges.

```python
# Extract total profit
total_profit = df['total_profit']

# Create histogram
plt.figure(figsize=(10,5))
plt.hist(total_profit, bins=5, color='purple', edgecolor='black')

# Labels and title
plt.xlabel("Profit Range")
plt.ylabel("Frequency")
plt.title("Distribution of Total Monthly Profit")

# Show plot
plt.show()
```

Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart.

```python
# Calculate total yearly sales for each product
total_facecream = df['facecream'].sum()
total_facewash = df['facewash'].sum()
total_toothpaste = df['toothpaste'].sum()
total_bathingsoap = df['bathingsoap'].sum()
total_shampoo = df['shampoo'].sum()
total_moisturizer = df['moisturizer'].sum()

# Prepare data for pie chart
products = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
sales = [total_facecream, total_facewash, total_toothpaste, total_bathingsoap, total_shampoo, total_moisturizer]

# Plot pie chart
plt.figure(figsize=(10,7))
plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=90)

# Title
plt.title("Total Sales Distribution for Each Product (Last Year)")

# Display chart
plt.show()
```

Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot.

```python
# Extract data
month = df['month_number']
bathingsoap = df['bathingsoap']
facewash = df['facewash']

# Create subplots
plt.figure(figsize=(12,6))

# Subplot 1: Bathing Soap
plt.subplot(2, 1, 1)
plt.plot(month, bathingsoap, marker='o', color='green')
plt.title("Bathing Soap Sales per Month")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.grid(True)

# Subplot 2: Face Wash
plt.subplot(2, 1, 2)
plt.plot(month, facewash, marker='o', color='blue')
plt.title("Face Wash Sales per Month")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

Exercise 10: Read all product sales data and show it using the stack plot.

```python
# Extract month numbers
month = df['month_number']

# Extract product sales
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

# Create stack plot
plt.figure(figsize=(12,6))

plt.stackplot(
    month,
    facecream,
    facewash,
    toothpaste,
    bathingsoap,
    shampoo,
    moisturizer,
    labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'],
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']
)

# Labels and title
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("All Product Sales Data - Stack Plot")

# Show legend
plt.legend(loc='upper left')

# Display plot
plt.show()
```
